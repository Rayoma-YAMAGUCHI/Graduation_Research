## 必要なモジュールをインポートする

```python
import pandas as pd
import kmapper as km
import matplotlib.pyplot as plt
import kmapper as km
from kmapper import jupyter # Creates custom CSS full-size Jupyter screen
import sklearn
from sklearn import manifold #MDSのライブラリimport
from IPython.display import IFrame
```

## スノーボードの使用するカタログスペック(csvファイル)を取り込む

```python
df = pd.read_csv("board_shape.csv", index_col=0)
df_board = df.drop(df.columns[[0]], axis=1)
df_board
```

```python
data = df_board.values
```

## Mapper解析

```python
# Initialize
mapper = km.KeplerMapper(verbose=1)

# Fit to and transform the data
projected_data = mapper.fit_transform(data, projection=[0,1]) # X-Y axis
#projected_data = mapper.fit_transform(data, projection=manifold.MDS(n_components=2, n_init=1, max_iter=100,random_state=0))
#projected_data = mapper.fit_transform(data, projection=manifold.MDS(n_components=2, random_state=0))
                                      
# Create a cover with 10 elements
cover = km.Cover(n_cubes=20, perc_overlap=0.5)

# Create dictionary called 'graph' with nodes, edges and meta-information
graph = mapper.map(projected_data, 
                   df_board, 
                   clusterer=sklearn.cluster.DBSCAN(eps=0.6, min_samples=3), 
                   cover=cover)

# Visualize it
_ = mapper.visualize(graph,
                     path_html="output/snowboard_output.html",
                     title="snowboard",
                     color_values=df.iloc[:,0],
                     color_function_name=["SHAPE"]
                     )

## Uncomment the below to view the above-generated visualization
#
#jupyter.display(path_html="output/snowboard_output.html")

## Alternatively, use an IFrame to display a vis with a set width and height
#
IFrame(src="output/snowboard_output.html", width=1800, height=1000)
```
